```python
# Car name, Fuel type, Body type, Transmission, car rating, make(company build), registerd state, model all help in determing the price
```

Encoding the fuel type

```python
df['fuel_type'].unique()
```

```python
fuel_mapping = {
    'petrol' : 1,
    'diesel' : 2,
    'petrol & cng' : 3,
    'petrol & lpg' : 4
}

df['fuel_type'] = df['fuel_type'].map(fuel_mapping).astype('int')
```

Encoding Body type

```python
df['body_type'].unique()
```

```python
body_mapping = {
    'hatchback' : 1,
    'sedan' : 2,
    'suv' : 3,
    'luxury sedan' : 4,
    'luxury suv' : 5
}

df['body_type'] = df['body_type'].map(body_mapping).astype('int')
```

Encoding Transmission

```python
df['transmission'].unique()
```

```python
trans_map = {
    'manual' : 1, 'automatic' : 2
}

df['transmission'] = df['transmission'].map(trans_map).astype('int')
```

Encoding car rating

```python
df['car_rating'].unique()
```

```python
rating_map = {
    'great' : 1, 'good' : 2, 'fair' : 3, 'overpriced' : 4
}

df['car_rating'] = df['car_rating'].map(rating_map).astype('int')
```

Encoding company

```python
df['make'].unique()
```

```python
make_map = {
    "maruti": 1,
    "hyundai": 2,
    "renault": 3,
    "honda": 4,
    "ford": 5,
    "nissan": 6,
    "toyota": 7,
    "tata": 8,
    "datsun": 9,
    "mahindra": 10,
    "volkswagen": 11,
    "audi": 12,
    "mercedes benz": 13,
    "chevrolet": 14,
    "bmw": 15,
    "skoda": 16,
    "volvo": 17,
    "fiat": 18,
    "isuzu": 19,
    "ssangyong": 20,
    "jeep": 21
}

df["make"] = df["make"].map(make_map).astype('int')
```

Encoding the state

```python
df['registered_state'].unique()
```

```python
state_mapping = {
    "delhi": 1,
    "uttar pradesh": 2,
    "haryana": 3,
    "maharashtra": 4,
    "karnataka": 5,
    "gujarat": 6,
    "jharkhand": 7,
    "punjab": 8,
    "telangana": 9,
    "andhra pradesh": 10,
    "west bengal": 11,
    "tamil nadu": 12,
    "uttarakhand": 13
}

df["registered_state"] = df["registered_state"].map(state_mapping).astype('int')
```

Encoding models name

```python
df['model'].unique().shape
```

```python
model_mapping = {model: i for i, model in enumerate(sorted(df["model"].unique()))}
df["model"] = df["model"].map(model_mapping)
```
